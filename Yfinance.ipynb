Fetching Stock Data
Goal: Get historical stock prices for one company.

```python
# Fetch Apple (AAPL) stock data
apple = yf.Ticker("AAPL")

# View historical prices for the past year
hist = apple.history(period="1y")
hist.head()  # shows the first 5 rows
```

Explanation:

Ticker("AAPL") → represents Apple stock

history(period="1y") → gets 1 year of daily stock prices

hist.head() → looks at the top rows of your data

Exploring the Data:
Goal: Understand what the stock data contains.:

```python
# Look at only the closing prices
close_prices = hist['Close']
print(close_prices.head())

# Look at the volume of shares traded
volume = hist['Volume']
print(volume.tail())  # last 5 rows

# Calculate a simple 20-day moving average (SMA)
hist['SMA_20'] = hist['Close'].rolling(window=20).mean()
hist[['Close','SMA_20']].tail()
```

Explanation:

Close → the daily closing price

Volume → number of shares traded

rolling(window=20).mean() → average over last 20 days

Mini Exercise:

Change the window size to 50 and see the difference.

Visualizing Stock Trends
Goal: Create charts to see trends clearly.

```python
# Plot closing prices
hist['Close'].plot(title="AAPL Closing Prices", figsize=(10,5))
plt.show()

# Plot closing price and 20-day moving average together
plt.figure(figsize=(10,5))
plt.plot(hist['Close'], label='Close')
plt.plot(hist['SMA_20'], label='20-day SMA')
plt.title("AAPL Closing Price & SMA")
plt.legend()
plt.show()
```
